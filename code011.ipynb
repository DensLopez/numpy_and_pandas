### ¿Qué pasa si mi archivo no está delimitado por coma?

el método tiene algunas funciones especiales

```python
#Leyendo archivo
#Usando delimitador "sep"
pd.read_csv('/home/dens/Dens/Codes/dataset/bestsellers.csv',sep=',')
```

Algunos no tienen encabezado así que hay que usar header
ponemos header=None

```python
#Leyendo archivo
#Usando delimitador "sep" y header = 0
df_books = pd.read_csv('/home/dens/Dens/Codes/dataset/bestsellers.csv',sep=',',header=0)
#si no tiene header le ponemos header=None
```
