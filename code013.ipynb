# Eliminar datos con Pandas

```python
import pandas as pd
import numpy as np
```

```python
#Usando Relative path
df_books = pd.read_csv('dataset/bestsellers.csv',sep=',',header=0)
df_books.head(5)    #Primeras 2 filas
```

### Eliminar columnas | drop columns
Referencia:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html


```python
#Drop de columnas
#Recordemos que axis=1 es columnas
df_books.drop('User Rating',axis=1).head(5)
```

```python
#Ejecutando para ver si lo borro del Data Frame
df_books.head(5)
```

Podemos ver que no lo borro, **drop()** solo lo borro de la salida

### Para borrarlo del Data Frame

```python
#Drop de columnas
#Recordemos que axis=1 es columnas
df_books.drop('Genre',axis=1,inplace=True)
```

```python
#Con el comando anterior borramos la columna Genre
df_books.head(5)
```

```python
#Otra forma de borrarlo haciendo esto
df_books = df_books.drop('Year',axis=1)
```

```python
df_books.head(5)
```

```python
#Borrando desde Python
del df_books['Price']
```

```python
#Usando Relative path
df_books = pd.read_csv('dataset/bestsellers.csv',sep=',',header=0)
df_books.head(5)    #Primeras 2 filas
```

### Eliminar filas | drop rows
Referencia:
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html


```python
#Borrando filas 
#Recordemos que axis=0 son filas
df_books.drop(0,axis=0).head(2)
```

```python
#Borrando por lista 
df_books.drop([0,1,3,5],axis=0).head(3)
```

```python
#borrando en un rango con una funcion
#Borrando del 0 al 5 con una Tupla
df_books.drop(range(0,6),axis=0).head(5)
```

### Recordemos que se puede aplicar el *Inplace* también a filas

## ¿Cómo agregar nuevas?

### Añadir columnas | Append columns


```python
#Imprimiendo dataFrame
df_books.head(3)
```

### Nan = Valor no Númerico

```python
#Agregando nueva columna
df_books['Nueva Columna'] = np.nan
```

```python
df_books.head(3)
```

```python
#Hay que saber de que tamaño es el Dataframe
fildf,coldf = df_books.shape
print(f'Filas = {fildf}\nColumnas = {coldf}')
```

Como se puede ver tenemos un frame de:
- 550 filas
- 8 columnas

```python
#Vamos a generar un vector de 0 a 550
data = np.arange(0,df_books.shape[0])
print(f'type(data) = {type(data)}')
```

```python
#Agregando nueva columna
df_books['Rango'] = data
```

```python
#El resultado es una nueva columna con los datos
df_books.head(3)
df_books.tail(3)
```

### Añadir filas | Append rows
Para las nuevas versiones de Pandas, el método para agregar filas fue cambiado a _append:

data_frame._append("list, tuple, dict", ignore*_*index=True)

O usa Concat:

pandas.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)

Referencias:
https://pandas.pydata.org/pandas-docs/version/1.4/reference/api/pandas.concat.html#pandas.concat


https://pandas.pydata.org/pandas-docs/version/1.4/reference/api/pandas.DataFrame.append.html


```python
#La mas veloz append
df_books._append(df_books)
```
