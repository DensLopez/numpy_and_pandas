## ¿Cómo crear un Excel?

#### Creando un set nulo para hacer un archivo en blanco de Excel para después cargar y modificarlo.
```python
import pandas as pd

# Creando datos nulos
data = {}
df_nulo = pd.DataFrame(data)

Nombre = 'Nombre.xlsx'

# Crea un archivo Excel con los parámetros dados
df_nulo.to_excel(Nombre, index=False)

# Cargar y modificar el archivo
with pd.ExcelWriter(Nombre, mode='a', engine='openpyxl') as writer:
    # Adjunta los DataFrames a las hojas existentes en el archivo
    df1 = pd.DataFrame({'A': [1, 2, 3]})
    df2 = pd.DataFrame({'B': [4, 5, 6]})
    df1.to_excel(writer, sheet_name='Sheet1', index=False)
    df2.to_excel(writer, sheet_name='Sheet2', index=False)

print("Datos escritos en", Nombre)
```

## Ajustar contenido de una hoja de Excel

```python
import pandas as pd
from openpyxl import load_workbook

# Nombre del archivo Excel
nombre_archivo = 'datos.xlsx'

# Cargar el archivo Excel
wb = load_workbook(nombre_archivo)

# Iterar sobre cada hoja del archivo Excel
for sheet_name in wb.sheetnames:
    # Obtener la hoja actual
    ws = wb[sheet_name]
    
    # Iterar sobre cada columna de la hoja actual
    for col in ws.columns:
        max_length = 0
        column = col[0].column_letter  # Obtener la letra de la columna
        for cell in col:
            try:  # Tratar de obtener la longitud del valor de la celda
                if len(str(cell.value)) > max_length:
                    max_length = len(str(cell.value))
            except:
                pass
        
        # Ajustar el ancho de la columna
        adjusted_width = (max_length + 2) * 1.2  # Ajuste adicional para mejor legibilidad
        ws.column_dimensions[column].width = adjusted_width

# Guardar los cambios en el archivo Excel
wb.save(nombre_archivo)

```
